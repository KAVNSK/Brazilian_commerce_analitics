# Olist Brazilian E-Commerce — рабочий план

**Датасет:** [Brazilian E-Commerce Public Dataset by Olist](https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce) (CC BY-NC-SA 4.0)  
**Локальные данные:** папка `data/` (CSV после скачивания с Kaggle).

Это короткий план без лишней теории. Он нужен, чтобы держать в голове последовательность шагов и не уходить в сторону.

Основной рабочий отчет: [`olist_case1_solution_report.ipynb`](olist_case1_solution_report.ipynb).

---

## 1. Минимальная структура данных

Для нашего кейса достаточно 5 таблиц:

| Файл | Зачем нужен |
|------|-------------|
| `olist_orders_dataset.csv` | даты и статус заказа |
| `olist_order_items_dataset.csv` | цена, фрахт и продавец |
| `olist_order_payments_dataset.csv` | тип оплаты и рассрочка |
| `olist_order_reviews_dataset.csv` | оценка клиента |
| `olist_customers_dataset.csv` | штат клиента |

Связи простые:
- `orders` -> `order_items`, `payments`, `reviews` по `order_id`;
- `orders` -> `customers` по `customer_id`.

Для демо-проекта этого более чем достаточно.

## 2. Какие метрики считаем в первую очередь

### Обязательные метрики
- `bad_review_rate`: доля заказов с `review_score <= 2`.
- `late_rate`: доля заказов, доставленных позже обещанной даты.
- `delay_days`: отклонение фактической даты от обещанной.

### Метрики для контекста
- `orders`: количество заказов.
- `GMV`: сумма `price + freight_value`.
- `AOV`: средний чек.

### Базовая статистика
- хи-квадрат: связь между `is_late` и `bad_review_flag`.

Этот набор короткий, но рабочий: его хватает, чтобы защитить кейс и дать бизнесу понятный вывод.

## 3. Порядок работы

1. Загрузить данные в DuckDB.
2. Проверить качество данных (пропуски дат, orphan-строки, статусы).
3. Собрать 2 витрины:
   - `v_order_facts` (общая картина),
   - `v_case_order_base` (кейс про late и bad).
4. Посчитать ключевые метрики (`bad_review_rate`, `late_rate`, `delay_days`).
5. Построить графики и проверить гипотезу.
6. Сформулировать короткий список действий для бизнеса.

Логика простая: сначала надежная база, потом выводы.

## 4. Структура рабочего ноутбука

Для этого кейса держим простую структуру:

1. Короткая постановка задачи.
2. Загрузка данных в DuckDB.
3. Проверка качества данных.
4. Витрина заказов и базовые метрики.
5. Витрина кейса (late/bad/seller) и risk-анализ.
6. Графики и базовые статпроверки.
7. Выводы и действия для бизнеса.

Именно по этой схеме собран `olist_case1_solution_report.ipynb`.

## 5. План визуализации (минимум для кейса)

1. Динамика заказов и GMV по месяцам.
2. Распределение `delivery_days`.
3. Доля плохих отзывов по группам задержки.
4. Scatter продавцов: `late_rate` vs `bad_review_rate`.
5. Распределение `review_score`.

Этих графиков достаточно, чтобы показать весь сюжет кейса без перегрузки.
